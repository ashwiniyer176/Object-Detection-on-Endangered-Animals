```python
import numpy as np
import pandas as pd
import os
from distutils.dir_util import copy_tree
import shutil
```

```python
def delete_folder(classnames,path):
    try:
        folders=os.listdir(path)
    except NotADirectoryError:
        return
    for c in range(len(folders)):
        if folders[c] not in classnames:
            try:
                shutil.rmtree(os.path.join(path,folders[c]))
            except NotADirectoryError:
                return
```

```python
def rename_one_folder(folder_path,classname):
    for idx,img in enumerate(os.listdir(folder_path)):
        try:
            if img=="Label":
                for i,d in enumerate(os.listdir(os.path.join(folder_path,"Label"))):
                    os.rename(os.path.join(folder_path,"Label",d),os.path.join(folder_path,"Label",f"{classname}_{i}.txt"))
            else:
                os.rename(os.path.join(folder_path,img),os.path.join(folder_path,f"{classname}_{idx}.jpg"))
        except FileExistsError:
            continue
```

```python
def remove_whitespace(folder_path):
    for i in range(len(os.listdir(folder_path))):
        os.rename(os.path.join(folder_path,folder_path[i]),os.path.join(folder_path,folder_path[i].replace(" ","_")))
```

```python
for i in ["train","test"]:
    try:
        for c in os.listdir(os.path.join(DIR_PATH,i)):
            print(f"{i} - {c}")
            rename_one_folder(os.path.join(DIR_PATH,i,c),c)
    except NotADirectoryError:
        break
```

```python
if os.path.exists("./data/annotations") is False:
    os.makedirs("./data/annotations/train")
    os.makedirs("./data/annotations/test")
    for i in ["train","test"]:
        NEW_PATH=os.path.join(DIR_PATH,i)
        print(NEW_PATH)
        for c in os.listdir(NEW_PATH):
            copy_tree(os.path.join(NEW_PATH,c,"Label"),os.path.join("./data/annotations",i))
```

```python
import re

def isnumeric(string):
    pattern=re.compile("[0-9.]+")
    if pattern.match(string) is None:
        return False
    return True

def create_csv(split_path):
    data=[]
    for x in os.listdir(split_path):
        filepath=os.path.join(split_path,x)
        file=open(filepath,"r")
        line=file.readline()
        line=line.replace("\n","")
        line=line.split(" ")
        line.append(filepath)
        print(line)
        for l in range(len(line)-1):
            if isnumeric(line[l]):
                line[l]=float(line[l])
        data.append(line)
    return data
```

```python
def copy_images_to_root(root_path):
    for i in os.listdir(root_path):
        classpath=os.path.join(root_path,i)
        for i,n in enumerate(os.listdir(classpath)):
            if n!="Label":
                shutil.copy(os.path.join(classpath,n),os.path.join(root_path,n))
```

```python
def find_height_and_width(patht):
    data=[]
    for i in os.listdir(patht):
        print(i)
        img=cv2.imread(os.path.join(patht,i))
        height,width,_=img.shape
        data.append([i,height,width])
    return data
```

```python
train_height=find_height_and_width("./data/train/")
test_height=find_height_and_width("./data/test/")
```

```python
train_height=pd.DataFrame(train_height,columns=["filename","height","weight"])
test_height=pd.DataFrame(test_height,columns=["filename","height","weight"])
```

```python
def splitter(x):
    print(x)
    return os.path.splitext(x.split("/")[-1])[0]
```
